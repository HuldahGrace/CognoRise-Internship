#### TASK-4: RED WINE QUALITY

#### IMPORTING LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### READING DATA

```python
df = pd.read_csv("cognoriseDatasets\winequality-red.csv")
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

#### DISTRIBUTION OF INPUT FEATURES

```python
fig, axs = plt.subplots(4,3, figsize=(7,7), layout="constrained")
fig.suptitle("Numerical Feature Visualization")
sns.boxplot(data=df, x='fixed acidity', ax=axs[0][0])
sns.boxplot(data=df, x='volatile acidity', ax=axs[0][1])
sns.boxplot(data=df, x='citric acid', ax=axs[0][2])
sns.boxplot(data=df, x='residual sugar', ax=axs[1][0])
sns.boxplot(data=df, x='chlorides', ax=axs[1][1])
sns.boxplot(data=df, x='free sulfur dioxide', ax=axs[1][2])
sns.boxplot(data=df, x='total sulfur dioxide', ax=axs[2][0])
sns.boxplot(data=df, x='density', ax=axs[2][1])
sns.boxplot(data=df, x='pH', ax=axs[2][2])
sns.boxplot(data=df, x='sulphates', ax=axs[3][0])
sns.boxplot(data=df, x='alcohol', ax=axs[3][1])
plt.show()
```

```python
df["quality"].agg(["unique","nunique"])
```

#### DISTRIBUTION OF QUALITY

```python
fig,axs=plt.subplots(figsize=(7,3))
sns.countplot(data=df,x="quality",color="lightseagreen")
```

#### IMPACT OF QUALITY BASED ON ALCOHOL

```python
fig=plt.figure(figsize=(7,3))
sns.boxplot(data=df,x="quality",y="alcohol")
```

#### RELATIONSHIP BETWEEN NUMERICAL FEATURES

```python
df_num = df.select_dtypes(include=["float"])
fig = plt.figure(layout="constrained",figsize=(10,5))
sns.heatmap(df_num.corr(), annot=True, linewidths=0.8)
```

#### IMPACT OF INPUT VARIABLES ON QUALITY OF RED WINE

```python
fig, axs = plt.subplots(4,3, figsize=(10,8), constrained_layout=True)
sns.boxenplot(data=df, x='quality', y='alcohol', ax=axs[0][0])
sns.boxenplot(data=df,y='sulphates', x='quality',  ax=axs[0][1])
sns.boxenplot(data=df,y='fixed acidity', x='quality',  ax=axs[0][2])
sns.boxenplot(data=df,y='volatile acidity', x='quality',  ax=axs[1][0])
sns.boxenplot(data=df,y='citric acid', x='quality',  ax=axs[1][1])
sns.boxenplot(data=df,y='residual sugar', x='quality',  ax=axs[1][2])
sns.boxenplot(data=df,y='chlorides', x='quality',  ax=axs[2][0])
sns.boxenplot(data=df,y='free sulfur dioxide', x='quality',  ax=axs[2][1])
sns.boxenplot(data=df,y='total sulfur dioxide', x='quality',  ax=axs[2][2])
sns.boxenplot(data=df,y='density', x='quality',  ax=axs[3][0])
sns.boxenplot(data=df,y='pH', x='quality',  ax=axs[3][1])
```

#### ALCOHOL VS PH ON QUALITY

```python
best_quality = df[df["quality"]>4]
```

```python
sns.scatterplot(data=best_quality,x="alcohol",y="pH",hue="quality")
```
